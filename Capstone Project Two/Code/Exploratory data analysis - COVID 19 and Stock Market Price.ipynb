```python
# Define function to calculate and display the list of correlated variables
def list_of_correlated_var(df):
    corr_matrix = df.drop('Adj Close', axis =1).corr().abs() #  Exclude Adjusted closing price - used as target variable
    high_corr_var=np.where(corr_matrix>0.8) # value more than 0.8 is considered to be correlated
    high_corr_var=[(corr_matrix.index[x],corr_matrix.columns[y]) for x,y in zip(*high_corr_var) if x!=y and x<y]

    #convert list of hihgly correlated to DataFrame
    list_of_hihgly_correlated_var = (pd.DataFrame(list(high_corr_var)))
    list_of_hihgly_correlated_var.columns = ['factor_one', 'factor_two']

    # For better readability, start index from one
    list_of_hihgly_correlated_var.index += 1
    return list_of_hihgly_correlated_var
```

```python
# call function on all variables -
list_of_correlated_var(data)
```

## Drop the Highly Correlated Variables

```python
# Correlation Matrix of response variable
corr_matrix = data.drop('Adj Close', axis =1).corr().abs()


# upper triangle of the correlation matrix 
upper = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(np.bool))


#variable to_drop, get the columns of our 'upper' variable, make sure the threshold is 0.80.
to_drop = [column for column in upper.columns if any(upper[column] > 0.80)]
```

```python
# Variable to be dropped! 
print('Variable to be dropped:',to_drop)
```

```python
# Drop the corelated variables
df_clean = data.drop(data[to_drop], axis=1)
```
